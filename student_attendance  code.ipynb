```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("/content/2018-2019_Daily_Attendance_20240429.csv")
```

```python
data['Date'] = pd.to_datetime(data['Date'], format='%Y%m%d')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Enrolled'], color='blue')
plt.title('Daily Trend of Enrolled Students')
plt.xlabel('Date')
plt.ylabel('Enrolled Students')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Absent'], bins=30, kde=True, color='red')
plt.title('Distribution of Absent Students')
plt.xlabel('Number of Absent Students')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data['Absent'], color='orange')
plt.title('Boxplot of Absent Students')
plt.xlabel('Number of Absent Students')
plt.grid(True)
plt.show()
```

```python
data['Month'] = data['Date'].dt.month
monthly_avg_enrolled = data.groupby('Month')['Enrolled'].mean()

plt.figure(figsize=(8, 6))
monthly_avg_enrolled.plot(kind='bar', color='purple')
plt.title('Monthly Average of Enrolled Students')
plt.xlabel('Month')
plt.ylabel('Average Enrolled Students')
plt.grid(True)
plt.show()
```

```python
data['Attendance Rate'] = (data['Present'] / data['Enrolled']) * 100

plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Attendance Rate'], color='blue')
plt.title('Daily Attendance Rate')
plt.xlabel('Date')
plt.ylabel('Attendance Rate (%)')
plt.grid(True)
plt.show()
```

```python
data['Enrolled Change'] = data['Enrolled'].diff()

plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Enrolled Change'], color='magenta')
plt.title('Daily Change in Enrolled Students')
plt.xlabel('Date')
plt.ylabel('Change in Enrolled Students')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Released'], bins=30, kde=True, color='brown')
plt.title('Distribution of Released Students')
plt.xlabel('Number of Released Students')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
rolling_mean = data['Enrolled'].rolling(window=30).mean()  # 30-day rolling mean
plt.plot(data['Date'], rolling_mean, color='green')
plt.title('Daily Trend of Enrolled Students (30-day Rolling Mean)')
plt.xlabel('Date')
plt.ylabel('Enrolled Students')
plt.grid(True)
plt.show()
```
